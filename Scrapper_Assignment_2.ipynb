```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.nyit.edu/engineering/faculty'  # Replace this with the URL of the website you want to scrape
response = requests.get(url)
print(response)
html_content = response.text

outputfile="Names_of_Professors.txt"
outputfile_1="bio_urls.txt"
outputfile_2="courses_taught.txt"
outputfile_3="bios.txt"

Parse = BeautifulSoup(html_content, 'html.parser') 
paragraphs= Parse.find_all('div', class_="media__body spacing-third")

# Extract the text from each paragraph
for paragraph in paragraphs:
    print(paragraph.get_text())
    outputfile= open("Names_of_Professors.txt","a")
    outputfile.write(paragraph.get_text())
    outputfile.write(" \n")
    outputfile.close()
    specific_header = paragraph.find('h2')
    
    if specific_header:
        # Extract links associated with the header
        header_text = specific_header.get_text().strip()
        print("Header Text:", header_text)
        outputfile_1= open("bio_urls.txt","a")
        L = header_text+ " \n"
        outputfile_1.write(L)
        # Check if the header has an anchor tag with 'href' attribute
        header_link = specific_header.find('a', href=True)
        if header_link:
            link_url = header_link['href']
            link_url2= "https://www.nyit.edu"+link_url
            print("Header Link:", link_url2)
            outputfile_1.write("Header Link:"+ link_url2+ " \n")
            
            linked_response = requests.get(link_url2)
            linked_html_content = linked_response.text

            # Parse the linked page's HTML content using BeautifulSoup
            linked_soup = BeautifulSoup(linked_html_content, 'html.parser')
            #print(linked_soup)
            linked_paragraphs = linked_soup.find('div', class_="article-inner border-top-right longform")
            
            #if linked_header_div:
            #     linked_paragraphs = linked_header_div.find_next_siblings('p')
            i=1
            for paragraph in linked_paragraphs:
                if i==2:
                    print("Linked Page Paragraph:", paragraph.get_text())
                    outputfile_3= open("bios.txt","a")
                    outputfile_3.write(paragraph.get_text())
                    outputfile_3.write(" \n")
                    outputfile_3.close()
                i=i+1
                    
            #if linked_header_div:
            headers =linked_soup.find_all('h3')
            #headers = linked_soup.find_all('h3')
            for header in headers:
                header_text = header.get_text().strip()
                #print(header_text)
                if header_text=="Courses Taught at New York Tech":
                    unordered_list = header.find_next('ul')
                    if unordered_list:
                        # Extract the list items from the unordered list
                        list_items = unordered_list.find_all('li')
                        data_below_unordered_list = [item.get_text().strip() for item in list_items]
                        print(data_below_unordered_list)
                        outputfile_2= open("courses_taught.txt","a")
                        outputfile_2.writelines(data_below_unordered_list)
                        outputfile_2.write(" \n")
                        outputfile_2.close()
                    else:
                        print("Courses are not mentioned")

        else:
            print("No Bio attached.")
            L= "Header Link:"+"No Bio attached."+ " \n"
            outputfile_1.write(L)
        
        outputfile_1.close()
    else:
        print("Specific header not found inside the div.")

#paragraphs = Parse.find_all('p')
# Extract the text from each paragraph
#for paragraph in paragraphs:
#    print(paragraph.get_text())
```
